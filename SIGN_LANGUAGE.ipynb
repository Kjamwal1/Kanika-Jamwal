```python
import os
from pathlib import Path

# 1. First, let's verify and fix the directory structure
def setup_yolov5_environment():
    # Create proper dataset structure
    dataset_path = Path('/content/yolov5/data')
    dataset_path.mkdir(exist_ok=True)

    # Create required subdirectories
    (dataset_path/'images'/'train').mkdir(parents=True, exist_ok=True)
    (dataset_path/'images'/'val').mkdir(parents=True, exist_ok=True)
    (dataset_path/'labels'/'train').mkdir(parents=True, exist_ok=True)
    (dataset_path/'labels'/'val').mkdir(parents=True, exist_ok=True)

    # Create proper data.yaml file
    data_yaml_content = f"""
path: /content/yolov5/data  # absolute path
train: images/train
val: images/val
nc: 3  # UPDATE with your number of classes
names: ['class1', 'class2', 'class3']  # UPDATE with your class names
"""

    with open(dataset_path/'data.yaml', 'w') as f:
        f.write(data_yaml_content)

    return dataset_path/'data.yaml'

# 2. Fix the Weights & Biases (wandb) warning
def disable_wandb():
    os.environ['WANDB_DISABLED'] = 'true'

# 3. Fix CUDA warnings (these are harmless but we can suppress them)
def suppress_cuda_warnings():
    os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
    os.environ['CUDA_LAUNCH_BLOCKING'] = '1'

# 4. Main execution
def main():
    # Apply all fixes
    suppress_cuda_warnings()
    disable_wandb()
    yaml_path = setup_yolov5_environment()

    # Verify paths
    print(f"\nVerifying paths...")
    print(f"data.yaml exists: {yaml_path.exists()}")
    print(f"train images exists: {(yaml_path.parent/'images'/'train').exists()}")

    # Run training with correct parameters
    !cd /content/yolov5 && python train.py \
        --img 640 \
        --batch 16 \
        --epochs 100 \
        --data "{yaml_path}" \
        --cfg models/custom_yolov5s.yaml \
        --weights yolov5s.pt \
        --name custom_training \
        --cache ram

if __name__ == '__main__':
    main()
```

```python
import os
from pathlib import Path

# 1. First, let's find the maximum class ID in your labels
def find_max_class_id(labels_dir):
    max_id = -1
    for label_file in Path(labels_dir).glob('*.txt'):
        with open(label_file) as f:
            for line in f:
                class_id = int(line.strip().split()[0])
                max_id = max(max_id, class_id)
    return max_id

# 2. Update data.yaml with correct number of classes
def update_data_yaml(data_yaml_path, num_classes):
    with open(data_yaml_path) as f:
        data = yaml.safe_load(f)

    data['nc'] = num_classes
    # Ensure class names list matches the number of classes
    if 'names' not in data or len(data['names']) != num_classes:
        data['names'] = [f'class_{i}' for i in range(num_classes)]

    with open(data_yaml_path, 'w') as f:
        yaml.dump(data, f, sort_keys=False)

# 3. Main execution
def main():
    # Paths to your data
    labels_dir = '/content/yolov5/train/labels'
    data_yaml_path = '/content/yolov5/data/data.yaml'

    # Find actual number of classes needed
    max_class_id = find_max_class_id(labels_dir)
    required_classes = max_class_id + 1  # classes are 0-indexed

    print(f"Max class ID found: {max_class_id}")
    print(f"Required number of classes (nc): {required_classes}")

    # Update data.yaml
    update_data_yaml(data_yaml_path, required_classes)

    # Verify update
    print("\nUpdated data.yaml contents:")
    with open(data_yaml_path) as f:
        print(f.read())

    # Now run training
    !cd /content/yolov5 && python train.py \
        --img 640 \
        --batch 16 \
        --epochs 100 \
        --data {data_yaml_path} \
        --cfg models/custom_yolov5s.yaml \
        --weights yolov5s.pt \
        --name fixed_training \
        --cache ram

if __name__ == '__main__':
    import yaml
    main()
```

```python
import os
import yaml
from pathlib import Path

# 1. Verify your dataset structure
dataset_path = Path('/content/tutorial-sign-main')
print("Current dataset structure:")
!tree {dataset_path} -L 3  # Show folder structure

# 2. Create proper data.yaml configuration
def create_data_yaml(dataset_path, class_names):
    yaml_content = f"""
# Dataset
path: {dataset_path}  # dataset root dir
train: images/train  # train images (relative to path)
val: images/val  # val images (relative to path)
test:  # optional test images

# Classes
nc: {len(class_names)}  # number of classes
names: {class_names}  # class names
"""

    yaml_path = dataset_path/'data.yaml'
    with open(yaml_path, 'w') as f:
        f.write(yaml_content)
    return yaml_path

# 3. Main execution
def main():
    # Define your class names (UPDATE THESE)
    class_names = ['stop', 'yield', 'speed_limit', 'warning', 'no_entry', 'pedestrian']  # EXAMPLE - CHANGE ME

    # Create data.yaml
    yaml_path = create_data_yaml(dataset_path, class_names)

    # Verify the file was created
    print("\nCreated data.yaml with contents:")
    with open(yaml_path) as f:
        print(f.read())

    # Check label class IDs
    print("\nChecking label files...")
    max_class_id = -1
    for split in ['train', 'val']:
        label_dir = dataset_path/f'labels/{split}'
        if label_dir.exists():
            for label_file in label_dir.glob('*.txt'):
                with open(label_file) as f:
                    for line in f:
                        class_id = int(line.strip().split()[0])
                        max_class_id = max(max_class_id, class_id)

    print(f"Maximum class ID found: {max_class_id}")
    print(f"Number of classes in data.yaml: {len(class_names)}")

    if max_class_id >= len(class_names):
        print("\n⚠️ WARNING: Some labels have class IDs >= number of classes!")
        print("Please either:")
        print("1. Update your class_names list to include all classes, or")
        print("2. Fix your label files to use correct class IDs (0-based index)")
        return

    # Start training
    print("\nStarting training...")
    !cd /content/yolov5 && python train.py \
        --img 640 \
        --batch 16 \
        --epochs 100 \
        --data "{yaml_path}" \
        --weights yolov5s.pt \
        --name sign_detection \
        --cache ram

if __name__ == '__main__':
    main()
```
